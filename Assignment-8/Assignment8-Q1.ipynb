```python
from matplotlib.patches import Rectangle
import matplotlib.pyplot as plt
import numpy as np
```

```python
ax = plt.gca()
r = Rectangle((0.1, 0.1), 0.5, 0.5, fill=False, edgecolor='r')
ax.add_patch(r)

def click(event):
    x = event.xdata
    y = event.ydata
    if event.inaxes:
        if 0.1 <= x <= 0.6 and 0.1<=y<=0.6:
            ax.text(x, y, 'o')
        else:
            ax.text(x, y, 'x')
    

plt.gcf().canvas.mpl_connect('button_press_event', click)
```
